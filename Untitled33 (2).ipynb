```python

# Read the Excel file
xls = pd.ExcelFile(file_path)

# List to store DataFrames of all sheets
dataframes = []

# Iterate through each sheet and read into DataFrames
for sheet_name in xls.sheet_names:
    df = xls.parse(sheet_name)
    dataframes.append(df)

# Concatenate the DataFrames vertically
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a CSV file
csv_output_path = "combined_data.csv"
combined_df.to_csv(csv_output_path, index=False)

print(f"Combined DataFrame saved to {csv_output_path}")
```

# which one month had highest number of sales

```python
df1=cleaned_df[cleaned_df[ 'Order Date']!='Order Date']
df1.head()
```

```python


# Convert "Order Date" column to datetime format
df1["Order Date"] = pd.to_datetime(df1["Order Date"])

# Extract the month and create a new column
df1["Order Month"] = df1["Order Date"].dt.strftime("%m")

# Save the DataFrame with the new column
cleaned_with_month_column_path = "cleaned_with_month_column.csv"
cleaned_df.to_csv(cleaned_with_month_column_path, index=False)

print(f"DataFrame with new column saved to {cleaned_with_month_column_path}")
```

```python
df1.head()
```

```python
# Converting to Numeric Values 
df1['Quantity Ordered']=pd.to_numeric(df1['Quantity Ordered']) 
df1['Price Each']=pd.to_numeric(df1['Price Each'])
```

```python
# Adding a new column for Sales Amount 
df1['Sales']=df1['Quantity Ordered']*df1['Price Each']
```

```python
df1.head()
```

```python
 # Groupby Month Feature and finding Total sales in each month 
result=df1.groupby("Order Month").sum() 
result
```

```python
import matplotlib.pyplot as plt
```

```python

# Group data by "Order Month" and calculate total sales
grouped_df = result.groupby("Order Month")["Sales"].sum().reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(grouped_df["Order Month"], grouped_df["Sales"])
plt.xlabel("Order Month")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.show()
```

```python
def get_state(address): 
    return address.split(",")[2].split(" ") 
get_state('917 1st St, Dallas, TX 75001')
```

```python
# Adding New Feature City with State code
def get_city(address):
    return address.split(",")[1] 
def get_state(address):
    return address.split(",")[2].split(" ")[1]
df1['City']=df1['Purchase Address'].apply(lambda x: get_city(x)+ ' '  + get_state(x))
df1.head()
    
```

```python
result1=df1.groupby("City").sum() 
result1
```

```python
# Group data by "Sales Month" and calculate total sales
grouped_df = result1.groupby("City")["Sales"].sum().reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(grouped_df["City"], grouped_df["Sales"])
plt.xlabel("City")
plt.ylabel("Sales")
plt.title("City Sales")
plt.xticks(rotation=45, ha="right")

# Display exact sales values on the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va="bottom")

plt.tight_layout()
plt.show()
```

```python
# Group data by "Product" and calculate total quantity ordered and average price
grouped_df = result3.groupby("Product").agg({"Quantity Ordered": "sum", "Price Each": "mean"}).reset_index()

# Create a figure and two axes
fig, ax1 = plt.subplots(figsize=(42, 15))
ax2 = ax1.twinx()

# Plot "Quantity Ordered" on ax1
ax1.bar(grouped_df["Product"], grouped_df["Quantity Ordered"], color="b", alpha=0.6)
ax1.set_xlabel("Product")
ax1.set_ylabel("Quantity Ordered", color="b")

# Plot "Price" on ax2
ax2.plot(grouped_df["Product"], grouped_df["Price Each"], color="r", marker="o")
ax2.set_ylabel("Price", color="r")
# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)


# Display the chart
plt.title("Quantity Ordered and Price by Product")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
